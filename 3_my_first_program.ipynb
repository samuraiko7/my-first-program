```python
# Creating my first program.
#
# I am trying to create an immersive roleplay experience that should feel both humorous and true to a dystopian sci-fi setting, while still fulfilling the 3 tasks given by Rauf.
# The program is emulating a conversation between a droid of the Empire and a simple working class citizen
# who's spent his whole life at the mines doing physical labour for the state.
# The droid's responses are condescending, with a strong authoritarian tone, making the user feel like a cog in the Empire's machine.


import time # To make the flow of the program feel more natural, dialouge-like.

print("Attention! You're hereby summoned for your complete and uninterrupted attention, citizen of the Galatic Empire!")
time.sleep(3)

print() # I found about this being a way to add and empty line.
print("You've been assigned a personal intelligence droid to ask you a few questions and provide some real-time feedback based on your answers.")
time.sleep(3)

while True: # To create a loop that breaks only when input includes "yes" or "no". PROBLEM: if input includes both yes and no, in whichever order, it will always count as "yes".
  print()
  consent = input("Do you agree to the collection of your data for this purpose? ")
  time.sleep(2)

  if "yes" in consent or "Yes" in consent:
    print()
    print("DROID: Cooperation, as expected from a loyal servant of the Empire.")
    break
  elif "no" in consent or "No" in consent:
    print()
    print('''DROID: Please, The Empire doesnâ€™t need permission from a puny miner. You smell like the deepest, dankest shafts of the underworld. Perhaps a shower would be more fitting than resistance?
       We shall proceed.''')
    time.sleep(4) # This line requires a longer sleep-time due to it's high length.
    break
  else:
    print()
    print("DROID: Your response is unclear. Choose wisely, as the Empire does not take kindly to hesitation.")
    time.sleep(2)
time.sleep(2)

print()
print("Question 1: How many cycles around the sun has your planet witnessed since you began to exist? ") # Since this is a more vague question to read, it gets more sleep-time. And because it never meant to be answered, it isn't even a real input.
time.sleep(5)

print()
print("DROID: My apologies, let me rephrase this for your simple brain to understand:", end=' ') # I tried to use the "end" function without \n, to place the input box on the same line as the question, but apparently colab doesn't like that, so I split the question and put the other half in the input inquiry.
age = int(input("how old are you? "))  # How to make it specify that it wants integer number only? And in case of string input , to ask again instead of giving error?
time.sleep(2)

if 13 <= age <= 19: # If age is between 13 and 19, the subject is considered a teenager.
    print()
    print("DROID: We can see that you are a teenager, ")
    occupation = input("are you currently a student aswell? ").lower() # Here I learned about the .lower() function and wanted to try it to avoid  having to use too many "or" logical operators in the "if" statements.
    time.sleep(2)

    if "yes" in occupation or "i am" in occupation:
        print()
        print("DROID: Huh, what a waste of prime labour force.")
    elif "no" in occupation or "not" in occupation:
        print()
        print("DROID: Great! No need for that, you know your purpose.")
else:
  print()
  print("DROID: Prime age to be serving your state!")
time.sleep(2)

print()
colour = input("Question 2: What is your favourite colour? ").lower()
time.sleep(2)

if colour != "blue":
  print()
  print("DROID: Interesting pick, although blue would've been much better.")
elif colour == "blue":
  print()
  print("DROID: Perfect - the only right choice!")
time.sleep(2)

print()
hours = int(input("Question 3: How many hours do you study per day? "))
time.sleep(2)

if  2 <= hours <= 10:
  print()
  print("DROID: What are you studying that much for? Trying to become an alchemist? All that time could be spent providing for the Empire instead.")
elif hours >= 10:
  print()
  print("DROID: That is UNACCEPTABLE! We shall double your workload at the mines. I will see to it.")
elif hours < 2:
  print()
  print("DROID: Perfect! All the more time for the good ol' pickaxe to hit the stone.")
time.sleep(3)

print()
print("DROID: That concludes the survey.")
time.sleep(3)

print()
print("Calculating your results...")
time.sleep(4)

print('''

THE RESULTS:

Intelligence ---- 3/10
Prestige -------- 1/10
Charm ----------- 2/10
Potential ------- 0/10

OBJECTIVE EVALUATION:

Your scores indicate you are STRICTLY unfit for anything greater than working at the mines.
Recommendation: DO NOT aim for anything higher than physical labor for the Empire. Report back to the mines at once, off to work!''')
```
